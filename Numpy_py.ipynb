## Array shape

Array  has an attribute shape which tells us that  how many dimensions it has how many elements every dimension has


```python
class MyArray:
    def __init__(self, data):
        self.data = data

    @property
    def shape(self):
        rows = len(self.data)         # Total rows (outer list length)
        cols = len(self.data[0])      # Total columns (inner list length)
        return (rows, cols)           # Return shape as a tuple

# Ab ek object banate hain
my_array = MyArray([[1, 2, 3], [4, 5, 6]])
print(my_array)

# 'shape' ko directly access karte hain bina () ke
print(my_array.shape)  # Output: (2, 3)
```

```python
import numpy as np

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(np.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True
Ak = np.array([[12, 17, 19],
               [24, 35, 47],
               [57, 89, 95]])

# Create a boolean array where True indicates a prime number
prime_mask = np.vectorize(is_prime)(Ak)

# Use np.where to get indices of prime numbers
prime_indices = np.where(prime_mask)

print("Prime mask (boolean array):")
print(prime_mask)
print("\nIndices of prime numbers:")
print(prime_indices)

# Extract prime numbers from the original array using the indices
prime_numbers = Ak[prime_indices]

print("\nPrime numbers in the array:")
print(prime_numbers)
```

# **Broadcasting**
agr dono ka dimension same ni hai tu hum dimansion brhay gy by adding one in front (1,1,3) asy kr k or phr 1 brhta jay gaa khud b khud hee
# Same Shape Ho Toh Operation Hojayega: Agar dono arrays ka shape bilkul same ho, toh woh seedha operation mein chalayenge aur har element ka operation apne corresponding element ke saath hoga. Isme kisi special rule ki zaroorat nahi.

# Ek Dimension Ka Size Agar 1 Ho Toh Expand Ho Sakta Hai: Agar dono arrays ke shapes different hain, toh bhi operation ho sakta hai lekin sirf tab jab chhoti array ka size kisi dimension mein 1 ho. NumPy us dimension ko "stretch" kar ke doosre array ke size ke barabar kar dega. Matlab, agar ek array ka size 1 hai, toh woh size automatically badh sakta hai.

# Shuru Se Shapes Compare Karna: Agar arrays ke dimensions ka number different hai (jaise ek 2D hai aur doosra 3D), toh chhoti array ke shape ke shuru mein 1 add kiya jata hai taake dono arrays ke dimensions equal ho jayein. Is tarah dono arrays ka dimension equal ho jaata hai aur compare ho sakte hain.

# Her Dimension Mein Size Ya Toh Same Ya Phir 1 Hona Chahiye: Har dimension mein ya toh dono arrays ka size same hona chahiye ya phir ek ka size 1 hona chahiye. Agar kisi bhi dimension mein ye rule break hota hai, toh broadcasting kaam nahi karegi aur ValueError aayega. **bold text**


```python
#Add a Constant to Each Element of a Matrix
# Ek 3x3 matrix banayein aur har element mein ek constant (for example, 5) add karne ke liye broadcasting use karen
matrix = np.arange(1,10).reshape(3,3)
constant=5
result=matrix+constant
print(result)
```

```python
# 2. Row-Wise Addition in a Matrix
# Ek 4x3 matrix aur ek 1D array create karen jo 3 elements ka ho (ek row ke equal). Phir, is 1D array ko matrix ke har row mein add karne ke liye broadcasting ka use karen
matrix= np.arange(0,12).reshape(4,3)
matrix2=np.arange(0,3).reshape(1,3)
print(matrix+matrix2)
```

```python
# Column-Wise Multiplication in a Matrix
# Ek 3x4 matrix aur ek 1D array banayein jo 3 elements ka ho (ek column ke equal). Phir, is array ko matrix ke har column mein multiply karne ke liye broadcasting ka use karen.
matrix=np.arange(0,12).reshape(3,4)
matrix1=np.arange(0,3).reshape(3,1)
print(matrix)
print(matrix1)
print(matrix*matrix1)
```

```python
# Outer Product of Two Vectors
# Do 1D arrays banayein (for example, [1, 2, 3] aur [4, 5, 6]). Inka outer product calculate karne ke liye broadcasting use karen, jisse ek 3x3 matrix mile jismein har element array1[i] * array2[j] ho.
arr1=np.arange(1,4)
arr2=np.arange(4,7)
array=np.outer(arr1,arr2)
print(array)
```
